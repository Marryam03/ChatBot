```python
import time
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
model_name = "microsoft/DialoGPT-medium"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
Name = input('Your Name: ')
print('Chat (Please ask a question to start a conversation, or type "quit" to exit): ')
# Initialize chat history
chat_history = None

while True:
    # Ask the user for input
    question = input('{}: '.format(Name))

    # Check if the user wants to quit
    if question.lower() == 'quit':
        print('Chatbot: Goodbye!')
        break

    # Encode the input
    user_input = tokenizer.encode(question + tokenizer.eos_token, return_tensors="pt")

    # Concatenate the user input with the chat history
    chatbot_input = torch.cat([chat_history, user_input], dim=-1) if chat_history is not None else user_input

    # Generate a response
    chat_history = model.generate(
        chatbot_input,
        max_length=1000,
        do_sample=True,
        top_k=70,
        top_p=0.95,
        temperature=0.7,
        pad_token_id=tokenizer.eos_token_id,
    )

    # Print the output
    chat = tokenizer.decode(chat_history[:, chatbot_input.shape[-1]:][0], skip_special_tokens=True)
    print(f"Chatbot: {chat}")
```
